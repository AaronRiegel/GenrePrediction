```python
import pandas as pd
import numpy as np
import json
from gensim.models import word2vec
import seaborn as sns
import matplotlib.pyplot as plt
import re
import nltk
import warnings
```

```python
warnings.filterwarnings('ignore')
nltk.download('stopwords')
```

## Data processing

Reading file from *movie_data.json*

```python
with open(r'movie_data.json', 'r') as json_file:  
    data = json.load(json_file)
```

```python
movies = pd.DataFrame(data)
```

```python
movies.drop(movies.columns.difference(['id', 'title', 'genres', 'release_date', 'overview']), axis=1, inplace=True)
movies.dropna(inplace=True)
movies = movies[['id', 'title', 'genres', 'release_date', 'overview']]
movies['release_date'] = pd.to_datetime(movies['release_date'])
movies.info()
```

Creating binary columns for genres

```python
def update_genre(col):
    for genre in col.genres:
        col.loc[genre['name']] = 1
    return col
```

```python
genres = set()
for movie in movies['genres']:
    for genre in movie:
        genres.add(genre['name'])
genres = list(genres)

movies = pd.merge(movies,
                  pd.DataFrame(data=np.zeros((movies.shape[0], len(genres)), dtype=int), columns=genres),
                  left_index=True,
                  right_index=True)
movies = movies.apply(update_genre, axis=1)
movies.drop('genres', inplace=True, axis=1)
movies.head()
```

## Visualization

```python
plt.figure(figsize=(12,6)) 
genre_count = movies[genres].sum().sort_values(ascending=False)
g = sns.barplot(genre_count.values, genre_count.index, palette='rainbow')
g.set_title('Genre distribution')
g.set_xlabel('Number of movies')
```

## Analysis

Clean Data

```python
def clean(text): # use regular expression to remove specific characters
    text = re.sub("\'", " ", text)
    text = re.sub("[^a-zA-Z]"," ",text)
    text = ' '.join(text.split())
    text = text.lower()
    return text
```

```python
movies['clean_overview'] = movies['overview'].astype(str).apply(lambda x: clean(x))
```

```python
movies.head()
```

```python
def word_counts(text, terms):
    all_words = ' '.join([t for t in text]) 
    all_words = all_words.split() 
    
    fdist = nltk.FreqDist(all_words) 
    words_df = pd.DataFrame({'word':list(fdist.keys()), 'count':list(fdist.values())}) 
  
  # selecting top 20 most frequent words 
  
    d = words_df.nlargest(columns="count", n = terms) 
  # visualize words and frequencies
    plt.figure(figsize=(20,30)) 
    ax = sns.barplot(data=d, x= "count", y = "word") 
    ax.set(ylabel = 'Word') 
    plt.show()
```

```python
word_counts(movies['clean_overview'], 200)
```

```python
stpwrds = set(nltk.corpus.stopwords.words('english'))
```

```python
def remove_stops(text):
    cleaned = [w for w in text.split() if not w in stpwrds]
    return ' '.join(cleaned)
```

```python
movies['clean_overview'] = movies['clean_overview'].astype(str).apply(lambda x: remove_stops(x))
```

```python
word_counts(movies['clean_overview'], 200)
```

## Vanilla SVM

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
def overlap_precision(y_test, predictions):
    """
    Returns a precision value of all the movies who have
    at least one genre overlap between the y_test and the prediction.
    """
    overlap = 0
    no_overlap = 0
    for index, row in enumerate(y_test):
        if np.any(row & predictions[index]):
            overlap += 1
        else:
            no_overlap += 1
    return overlap / (no_overlap + overlap)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectorizer = TfidfVectorizer(max_df=0.8, max_features=10000)
```

```python
X = movies['clean_overview']
y = movies[genres]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
tfidf_xtrain = tfidf_vectorizer.fit_transform(X_train)
tfidf_xtest = tfidf_vectorizer.transform(X_test)
```

```python
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier
```

```python
model = OneVsRestClassifier(SVC(gamma = 'auto',kernel = 'linear'))
```

Fitting and predictions takes up to one minute

```python
model.fit(tfidf_xtrain, y_train)
predictions = model.predict(tfidf_xtest)
```

```python
print(classification_report(predictions,y_test))
print('Overlap precision: {:0.2f}'.format(overlap_precision(y_test, predictions)))
```

## Multi Label Binarization

```python
def make_genre_list(s):
    gen = []
    for g in genres:
        if s.loc[g] == 1:
            gen.append(g)
    return gen
```

```python
movies['genre_list'] = movies.apply(lambda row : make_genre_list(row), axis = 1)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
multilabel_binarizer = MultiLabelBinarizer()
multilabel_binarizer.fit(movies['genre_list'])

y = multilabel_binarizer.transform(movies['genre_list'])
X = movies['clean_overview']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
tfidf_xtrain = tfidf_vectorizer.fit_transform(X_train)
tfidf_xtest = tfidf_vectorizer.transform(X_test)
```

```python
model = OneVsRestClassifier(SVC(gamma = 'auto', kernel = 'linear'))
```

Fitting and prediction can take up to one minute

```python
model.fit(tfidf_xtrain, y_train)
mbl_predictions = model.predict(tfidf_xtest)
```

```python
print(classification_report(mbl_predictions,y_test))
print('Overlap precision: {:0.2f}'.format(overlap_precision(y_test, mbl_predictions)))
```

```python
def get_poster_path(poster_path):
    return r'https://image.tmdb.org/t/p/w500' + str(poster_path)
```

```python
def save_image(col):
    try:
        urllib.request.urlretrieve(get_poster_path(col['poster_path']), r'../Data/Images/' + str(col['id']) + '.jpg')
    except:
        print(col)
```

```python
df = pd.DataFrame(data=zeros, columns=genres)
```

```python
movies = pd.merge(movies, df, left_index=True, right_index=True)
```

```python
def update_genre(col):
    for genre in col.genres:
        col.loc[genre['name']] = 1
        # print(col.loc[genre['name']])
    return col
        
```

```python
movies = movies.apply(update_genre, axis=1)
```
